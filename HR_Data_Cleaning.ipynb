```python
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
# load the dataset
df = pd.read_csv('HRDataset_v14.csv')
pd.options.display.max_columns = None
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

* DateofTermination has 207 missing values. This could suggest that some employees are still active or the termination dates are missing for certain records.
* ManagerID has 8 missing values, indicating that some employees do not have an assigned manager or this data is incomplete for a few records.

```python
df[df.ManagerID.isnull()][['ManagerID', "ManagerName", 'Department']]
```

* The ManagerName for all the rows with missing ManagerID is "Webster Butler". This suggests that the ManagerID for these employees may not have been recorded in the dataset, or perhaps there's a mismatch or inconsistency in how ManagerIDs are assigned for employees in the Production department.

```python
df[df.ManagerName== 'Webster Butler'][['ManagerID', "ManagerName", 'Department']]
```

* It appears that Webster Butler has a ManagerID of 39.0 in some records, but in others, the ManagerID is missing (NaN). This inconsistency may be due to incomplete or incorrect data entry for certain employees in the Production department.

```python
df['ManagerID'].fillna(39, inplace = True)
```

* Since Webster Butler has a ManagerID of 39, I have filled the missing ManagerID values (for rows where the manager is Webster Butler) with 39.0. This would ensure consistency and correct assignment of the manager to these employees.

```python
df.ManagerID = df.ManagerID.astype('int') 
```

```python

date_columns = ['DOB', 'DateofHire', 'DateofTermination', 'LastPerformanceReview_Date']
df[date_columns].dtypes
```

```python
# I changed the date colunms data type to datecolumns
for i in date_columns:
    df[i] = df[i].astype('datetime64[ns]')
```

```python
# Rearrange and format
df['Employee_Name'] = (df['Employee_Name'].str.split(',', expand = True).iloc[:,1] 
                    + ' '
                    + df['Employee_Name'].str.split(',', expand = True).iloc[:,0])
```

* Rearrange and format the Employee_Name column, where the names were written as Last Name, First Name, with a comma separating the last name and first name to First Name Last Name

```python
# Dropped redundant columns
df.drop(columns = ['MarriedID','MaritalStatusID', 'GenderID', 
                   'Termd', 'PerfScoreID', 'EmpStatusID',
                   'PositionID',], inplace = True)
```

```python
# reorder of columns
cols_reorder = ['EmpID', 'Employee_Name','Sex', 'MaritalDesc', 'DeptID',
                'FromDiversityJobFairID', 'Salary',
                'Position', 'State', 'Zip', 'DOB', 
                'CitizenDesc', 'HispanicLatino', 'RaceDesc', 'DateofHire',
                'DateofTermination', 'TermReason', 'EmploymentStatus', 'Department',
                'ManagerName', 'ManagerID', 'RecruitmentSource', 'PerformanceScore',
                'EngagementSurvey', 'EmpSatisfaction', 'SpecialProjectsCount',
                'LastPerformanceReview_Date', 'DaysLateLast30', 'Absences'
       ]
```

```python
df = df[cols_reorder]
df.head()
```

```python
# saving cleaned data
df.to_csv('HR_Employee_Cleaned_Data.csv', index = False)
```
